```
from bs4 import BeautifulSoup
import requests
import pandas as pd
import streamlit as st
```

```
Username = input('Enter a github Username: ')
```

```
auth = requests.get('https://github.com/'+Username)
print(auth.content)
```

```
page = BeautifulSoup(auth.content,'html.parser')
info = {}
```

```
page
```
